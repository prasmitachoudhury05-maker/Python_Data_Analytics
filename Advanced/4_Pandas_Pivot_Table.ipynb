```python
#Importing Libraries
import pandas as pd
!pip install datasets
from datasets import load_dataset

import matplotlib.pyplot as plt
```

```python
#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
#Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Count of Jobs


```python
df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
df.groupby('job_title_short').size()
```

```python
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index

df_jobs_country = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

df_jobs_country_salary = df_jobs_country.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_jobs_country_salary = df_jobs_country_salary[job_titles]

df_jobs_country_salary.plot(kind='bar')
plt.ylabel('Median Salary (USD)')
plt.xlabel('')
plt.title('Median Salary by country and by Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
